STEAM_GAME

```python

# Crear una lista para almacenar los objetos JSON
datos_json = []

# Abre el archivo JSON usando jsonlines
with jsonlines.open('output_steam_games.json') as archivo_json:
    for objeto_json in archivo_json:
        datos_json.append(objeto_json)

# Convertir la lista de objetos JSON en un DataFrame de Pandas
df_steam_games = pd.DataFrame(datos_json)

df_steam_games
```

```python
df_steam_games.to_csv('steam_games.csv', index=False)
```

USER_REVIEWS

```python
def parse (path):
    with gzip.open(path, 'r') as g:
        for l in g:
            yield eval(l)
```

```python
rows = []
for line in parse('user_reviews.json.gz'):
    row = pd.json_normalize(line)
    rows.append(row)
```

```python
df = pd.concat(rows, ignore_index=True)
```

```python
df
```

```python

# Aplanar la columna "reviews" y expandir sus valores en nuevas columnas
df = df.explode("reviews")

# Convertir la columna "reviews" a un DataFrame
df_reviews = df["reviews"].apply(pd.Series)

# Unir el DataFrame original con el nuevo DataFrame de reseñas
df = pd.concat([df, df_reviews], axis=1)

# Eliminar la columna original "reviews" si es necesario
df.drop("reviews", axis=1, inplace=True)

# A continuación, puedes renombrar las columnas según lo que necesites
# Por ejemplo, si deseas renombrar la columna "review" a "review_text"
#df.rename(columns={"review": "review_text"}, inplace=True)

# Ahora tendrás un DataFrame donde cada fila contiene la información del usuario
# y los distintos valores de reseñas en columnas separadas
```

```python
df_reviews = df
df_reviews
```

```python
df_reviews.to_csv('user_reviews.csv', index=False)
```
